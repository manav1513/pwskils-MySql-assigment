```python
# Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

# DML stands for Data Manipulation Language, which is a subset of SQL (Structured Query Language) used to modify the data stored in a database. DML commands include INSERT, UPDATE, and DELETE. Here are explanations and examples for each:

# INSERT: The INSERT statement is used to add new rows of data into a table. It specifies the table name and the values to be inserted.
# Example: Suppose you have a table named "employees" with columns "id", "name", and "salary". To insert a new employee with an id of 101, a name of "John Doe", and a salary of 50000, you would use the following SQL statement:

    
# INSERT INTO employees (id, name, salary)
# VALUES (101, 'John Doe', 50000);


# UPDATE: The UPDATE statement is used to modify existing data in a table. It specifies the table name, the column(s) to be updated, and the new values.
# Example: Suppose you want to give John Doe a raise and update his salary to 55000. You would 

# UPDATE employees
# SET salary = 55000
# WHERE id = 101;


# DELETE: The DELETE statement is used to remove rows of data from a table. It specifies the table name and a condition that determines which rows to delete.
# Example: Suppose John Doe decides to leave the company and you need to remove his employee record. You would use the following SQL statement:


# DELETE FROM employees
# WHERE id = 101;
# This would remove the row where the "id" column equals 101, which is the row for John Doe.
```

```python
# Q4. What is DQL? Explain SELECT with an example.

# DQL stands for Data Query Language, which is a subset of SQL (Structured Query Language) used to retrieve data from a database. The most commonly used DQL command is SELECT, which retrieves data from one or more tables in the database. Here is an explanation and example of SELECT:

# SELECT: The SELECT statement is used to retrieve data from one or more tables in a database. It specifies the columns to be retrieved and the table(s) to retrieve them from. Additionally, it can include various clauses to filter, sort, and group the data.

# Example: Suppose you have a table named "employees" with columns "id", "name", "age", and "salary". To retrieve the names and ages of all employees with a salary greater than 50000, you would use the following SQL statement:


# SELECT name, age
# FROM employees
# WHERE salary > 50000;
```

```python
# Q7. Give the order of execution of SQL clauses in an SQL query.

# The order of execution of SQL clauses in an SQL query is as follows:

# FROM: The FROM clause specifies the table or tables that the query will be executed against. This is the first clause executed in the query.

# JOIN: The JOIN clause is used to join multiple tables together based on a specified condition. This clause is executed after the FROM clause.

# WHERE: The WHERE clause is used to filter the results of the query based on a specified condition. This clause is executed after the JOIN clause.

# GROUP BY: The GROUP BY clause is used to group the results of the query by one or more columns. This clause is executed after the WHERE clause.

# HAVING: The HAVING clause is used to filter the results of the query based on a condition applied to a group. This clause is executed after the GROUP BY clause.

# SELECT: The SELECT clause is used to specify the columns that will be returned in the result set. This clause is executed after all the previous clauses.

# DISTINCT: The DISTINCT keyword is used to remove duplicates from the result set. This clause is executed after the SELECT clause.

# ORDER BY: The ORDER BY clause is used to sort the result set by one or more columns. This clause is executed after the SELECT and DISTINCT clauses.

# LIMIT/OFFSET: The LIMIT and OFFSET clauses are used to limit the number of rows returned by the query and to specify an offset from the beginning of the result set. These clauses are executed after all the previous clauses.
```
